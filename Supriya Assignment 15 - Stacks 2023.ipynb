```python
#<aside>
ðŸ’¡ **Question 1**

Given an array **arr[ ]** of size **N** having elements, the task is to find the next greater element for each element 
of the array in order of their appearance in the array.Next greater element of an element in the array is the nearest 
element on the right which is greater than the current element.If there does not exist next greater of current element,
then next greater element for current element is -1. For example, next greater of the last element is always -1.

**Example 1:**
                                                                                                                                                                                                           
Input:
N = 4, arr[] = [1 3 2 4]
Output:
3 4 4 -1
Explanation:
In the array, the next larger element
to 1 is 3 , 3 is 4 , 2 is 4 and for 4 ?
since it doesn't exist, it is -1.

Example 2: 
Input:
N = 5, arr[] [6 8 0 1 3]
Output:
8 -1 1 3 -1
Explanation:
In the array, the next larger element to
6 is 8, for 8 there is no larger elements
hence it is -1, for 0 it is 1 , for 1 it
is 3 and then for 3 there is no larger
element on right and hence -1.

</aside>

"""To find the next greater element for each element in the array, you can use the following algorithm:

   1. Create an empty stack and an empty result array to store the next greater elements.
   2. Iterate through the array from right to left.
   3. For each element arr[i], perform the following steps:
      a. While the stack is not empty and the top element of the stack is smaller than or equal to arr[i], pop 
         elements from the stack.
      b. If the stack becomes empty, it means there is no greater element to the right of arr[i], so store -1 in
         the result array at index i.
      c. If the stack is not empty, store the top element of the stack in the result array at index i.
      d. Push arr[i] onto the stack.
   4. Return the result array."""

#Here's the implementation in Python:

def findNextGreaterElements(arr):
    stack = []
    result = [-1] * len(arr)
    
    for i in range(len(arr)-1, -1, -1):
        while stack and stack[-1] <= arr[i]:
            stack.pop()
            
        if stack:
            result[i] = stack[-1]
            
        stack.append(arr[i])
        
    return result
arr1 = [1, 3, 2, 4]
result1 = findNextGreaterElements(arr1)
print(result1)  # Output: [3, 4, 4, -1]

arr2 = [6, 8, 0, 1, 3]
result2 = findNextGreaterElements(arr2)
print(result2)  # Output: [8, -1, 1, 3, -1]
```

```python
#<aside>
ðŸ’¡ **Question 2**

Given an array **a** of integers of length **n**, find the nearest smaller number for every element such that the 
smaller element is on left side.If no small element present on the left print -1.

**Example 1:**

                                                                                                                                                                                                                  Input: n = 3
a = {1, 6, 2}
Output: -1 1 1
Explaination: There is no number at the
left of 1. Smaller number than 6 and 2 is 1.                                                                                                                                        Example 2:                                                                                                                                                                                                     Input: n = 6
a = {1, 5, 0, 3, 4, 5}
Output: -1 1 -1 0 3 4
Explaination: Upto 3 it is easy to see
the smaller numbers. But for 4 the smaller
numbers are 1, 0 and 3. But among them 3
is closest. Similary for 5 it is 4.

</aside>

"""To find the nearest smaller number for each element in the array, you can use the following algorithm:

   1. Create an empty stack and an empty result array to store the nearest smaller numbers.
   2. Iterate through the array from left to right.
   3. For each element a[i], perform the following steps:
      a. While the stack is not empty and the top element of the stack is greater than or equal to a[i], pop elements
         from the stack.
      b. If the stack becomes empty, it means there is no smaller element to the left of a[i], so store -1 in the
         result array at index i.
      c. If the stack is not empty, store the top element of the stack in the result array at index i.
      d. Push a[i] onto the stack.
      
   4. Return the result array."""

#Here's the implementation in Python:

def findNearestSmallerElements(a):
    stack = []
    result = [-1] * len(a)
    
    for i in range(len(a)):
        while stack and stack[-1] >= a[i]:
            stack.pop()
            
        if stack:
            result[i] = stack[-1]
            
        stack.append(a[i])
        
    return result
a1 = [1, 6, 2]
result1 = findNearestSmallerElements(a1)
print(result1)  # Output: [-1, 1, 1]

a2 = [1, 5, 0, 3, 4, 5]
result2 = findNearestSmallerElements(a2)
print(result2)  # Output: [-1, 1, -1, 0, 3, 4]
```

```python
#<aside>
ðŸ’¡ **Question 3**

Implement a Stack using two queues **q1** and **q2**.

**Example 1**

</aside>                                                                                                                                                                                                       Input:
push(2)
push(3)
pop()
push(4)
pop()
Output:3 4
Explanation:
push(2) the stack will be {2}
push(3) the stack will be {2 3}
pop()   poped element will be 3 the
        stack will be {2}
push(4) the stack will be {2 4}
pop()   poped element will be 4


Example 2:                                                                                                                                                                                                    Input:
push(2)
pop()
pop()
push(3)
Output:2 -1

</aside>

"""To implement a stack using two queues, you can follow the approach where one queue is used for the main operations
   of the stack (push and pop), while the other queue is used for auxiliary operations."""

#Here's the implementation in Python:

class Stack:
    def __init__(self):
        self.q1 = []
        self.q2 = []

    def push(self, val):
        self.q1.append(val)

    def pop(self):
        if not self.q1:
            return -1

        while len(self.q1) > 1:
            self.q2.append(self.q1.pop(0))

        popped_val = self.q1.pop(0)

        self.q1, self.q2 = self.q2, self.q1

        return popped_val
stack1 = Stack()
stack1.push(2)
stack1.push(3)
print(stack1.pop())  # Output: 3
stack1.push(4)
print(stack1.pop())  # Output: 4

stack2 = Stack()
stack2.push(2)
print(stack2.pop())  # Output: 2
print(stack2.pop())  # Output: -1
stack2.push(3)
```

```python
#<aside>
ðŸ’¡ **Question 4**

You are given a stack **St**. You have to reverse the stack using recursion.

**Example 1:**

</aside>                                                                                                                                                                                                            Input:St = {3,2,1,7,6}
Output:{6,7,1,2,3}  
    
Example 2:                                                                                                                                                                                                     Input:St = {4,3,9,6}
Example 2:                                                                                                                                                                                                     Input:St = {4,3,9,6}
Output:{6,9,3,4}

</aside> 

"""To reverse a stack using recursion, you can use the following approach:

   1. Create a recursive function called reverseStack that takes the stack as input.
   2. If the stack is empty or contains only one element, return the stack as it is.
   3. Otherwise, pop an element from the stack using the pop() operation.
   4. Recursively call reverseStack on the remaining stack.
   5. After the recursion returns, push the popped element back onto the stack using the push() operation.
   6. Return the reversed stack."""

#Here's the implementation in Python:

def reverseStack(stack):
    if len(stack) <= 1:
        return stack
    
    top = stack.pop()
    stack = reverseStack(stack)
    stack.append(top)
    
    return stack
stack1 = [3, 2, 1, 7, 6]
reversed_stack1 = reverseStack(stack1)
print(reversed_stack1)  # Output: [6, 7, 1, 2, 3]

stack2 = [4, 3, 9, 6]
reversed_stack2 = reverseStack(stack2)
print(reversed_stack2)  # Output: [6, 9, 3, 4]
```

```python
#<aside>
ðŸ’¡ **Question 5**

You are given a string **S**, the task is to reverse the string using stack.

**Example 1:**
Input: S="GeeksforGeeks"     
Output: skeeGrofskeeG

</aside> 

"""To reverse a string using a stack, you can follow the following approach:

   1. Create an empty stack.
   2. Iterate through each character in the string S.
   3. Push each character onto the stack.
   4. After iterating through all characters, create an empty string to store the reversed string.
   5. Pop each character from the stack and append it to the reversed string.
   6. The reversed string will be the desired output."""

#Here's the implementation in Python:

def reverseString(S):
    stack = []
    
    # Push each character onto the stack
    for char in S:
        stack.append(char)
    
    reversed_string = ""
    
    # Pop each character from the stack and append to the reversed string
    while stack:
        reversed_string += stack.pop()
    
    return reversed_string
S = "GeeksforGeeks"
reversed_string = reverseString(S)
print(reversed_string)  # Output: skeeGrofskeeG
```

```python
#<aside>
ðŸ’¡ **Question 6**

Given string **S** representing a postfix expression, the task is to evaluate the expression and find the final value. 
Operators will only include the basic arithmetic operators like ***, /, + and -**.

**Example 1:**

Input: S = "231*+9-"
Output: -4
Explanation:
After solving the given expression,
we have -4 as result.

Example 2:   
    
Input: S = "123+*8-"
Output: -3
Explanation:
After solving the given postfix
expression, we have -3 as result.

</aside>

"""To evaluate a postfix expression and find the final value, you can use a stack to perform the calculations.

   Here's the step-by-step approach to evaluate a postfix expression:

   1. Create an empty stack.
   2. Iterate through each character in the string S.
   3. If the character is a digit, convert it to an integer and push it onto the stack.
   4. If the character is an operator (+, -, *, /), pop the top two elements from the stack.
      . Perform the corresponding arithmetic operation on the popped elements.
      . Push the result back onto the stack.
   5. After iterating through all characters, the top element of the stack will be the final value."""

#Here's the implementation in Python:

def evaluatePostfix(S):
    stack = []
    
    for char in S:
        if char.isdigit():
            stack.append(int(char))
        else:
            operand2 = stack.pop()
            operand1 = stack.pop()
            
            if char == '+':
                result = operand1 + operand2
            elif char == '-':
                result = operand1 - operand2
            elif char == '*':
                result = operand1 * operand2
            elif char == '/':
                result = operand1 / operand2
            
            stack.append(result)
    
    return stack.pop()
S1 = "231*+9-"
result1 = evaluatePostfix(S1)
print(result1)  # Output: -4

S2 = "123+*8-"
result2 = evaluatePostfix(S2)
print(result2)  # Output: -3
```

```python
#<aside>
ðŸ’¡ **Question 7**

Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

Implement the `MinStack` class:

- `MinStack()` initializes the stack object.
- `void push(int val)` pushes the element `val` onto the stack.
- `void pop()` removes the element on the top of the stack.
- `int top()` gets the top element of the stack.
- `int getMin()` retrieves the minimum element in the stack.

You must implement a solution with `O(1)` time complexity for each function.

**Example 1:**

Input
["MinStack","push","push","push","getMin","pop","top","getMin"]
[[],[-2],[0],[-3],[],[],[],[]]

Output
[null,null,null,null,-3,null,0,-2]

Explanation
MinStack minStack = new MinStack();
minStack.push(-2);
minStack.push(0);
minStack.push(-3);
minStack.getMin(); // return -3
minStack.pop();
minStack.top();    // return 0
minStack.getMin(); // return -2

</aside>

"""To design a stack that supports push, pop, top, and retrieving the minimum element in constant time, you can use 
   an additional stack to keep track of the minimum elements.

   Here's the step-by-step approach to implement the MinStack class:

   1. Create two stacks: main_stack and min_stack.
      . main_stack will store all the elements pushed onto the stack.
      . min_stack will store the minimum elements encountered so far.
      
   2. When pushing an element onto the stack:
      . Push the element onto main_stack.
      . If min_stack is empty or the element is smaller than or equal to the top element of min_stack, push the 
        element onto min_stack.
        
   3. When popping an element from the stack:
      . Pop the top element from main_stack.
      . If the popped element is equal to the top element of min_stack, pop the top element from min_stack.
      
   4. When getting the top element of the stack, return the top element of main_stack.
   
   5. When retrieving the minimum element from the stack, return the top element of min_stack."""

#Here's the implementation in Python:

class MinStack:
    def __init__(self):
        self.main_stack = []
        self.min_stack = []

    def push(self, val):
        self.main_stack.append(val)

        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)

    def pop(self):
        if self.main_stack[-1] == self.min_stack[-1]:
            self.min_stack.pop()

        self.main_stack.pop()

    def top(self):
        return self.main_stack[-1]

    def getMin(self):
        return self.min_stack[-1]
minStack = MinStack()
minStack.push(-2)
minStack.push(0)
minStack.push(-3)
print(minStack.getMin())  # Output: -3
minStack.pop()
print(minStack.top())     # Output: 0
print(minStack.getMin())  # Output: -2
```

```python
#<aside>
ðŸ’¡ **Question 8**

Given `n` non-negative integers representing an elevation map where the width of each bar is `1`, compute how much
water it can trap after raining.

**Example 1:**
Input: height = [0,1,0,2,1,0,1,3,2,1,2,1]
Output: 6
Explanation: The above elevation map (black section) is represented by array [0,1,0,2,1,0,1,3,2,1,2,1]. In this case,
6 units of rain water (blue section) are being trapped.

Example 2:
Input: height = [4,2,0,3,2,5]
Output: 9

</aside>  

"""To compute the amount of water that can be trapped after raining on an elevation map represented by an array, you 
   can use a two-pointer approach.

   Here's the step-by-step approach to solve this problem:

   1. Initialize two pointers, left and right, pointing to the first and last elements of the array.
   
   2. Initialize two variables, left_max and right_max, to keep track of the maximum height encountered from the left 
      and right side respectively. Set both variables to 0.
      
   3. Initialize a variable water to store the total amount of trapped water, initially set to 0.
   
   4. While left is less than or equal to right, do the following:
      a. If height[left] is less than or equal to height[right]:
         . If height[left] is greater than left_max, update left_max to height[left].
         . Otherwise, calculate the amount of water that can be trapped at the current position left using the formula:
           water += left_max - height[left].
         . Increment left by 1.
      b. If height[left] is greater than height[right]:
         . If height[right] is greater than right_max, update right_max to height[right].
         . Otherwise, calculate the amount of water that can be trapped at the current position right using the formula: 
           water += right_max - height[right].
         . Decrement right by 1.

   5. After the loop, return the value of water."""

#Here's the implementation in Python:

def trap(height):
    left = 0
    right = len(height) - 1
    left_max = right_max = water = 0

    while left <= right:
        if height[left] <= height[right]:
            if height[left] > left_max:
                left_max = height[left]
            else:
                water += left_max - height[left]
            left += 1
        else:
            if height[right] > right_max:
                right_max = height[right]
            else:
                water += right_max - height[right]
            right -= 1

    return water
height1 = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
water1 = trap(height1)
print(water1)  # Output: 6

height2 = [4, 2, 0, 3, 2, 5]
water2 = trap(height2)
print(water2)  # Output: 9
```
